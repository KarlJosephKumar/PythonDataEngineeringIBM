```python
import requests
import pandas as pd
```

```python
#Enter your access key in the url
url = "http://api.exchangeratesapi.io/v1/latest?base=EUR&access_key=*******"  #Make sure to change ******* to your API key.
r = requests.get(url)
#Check status code of get request
r.status_code
```

```python
#Turning the text of the request to pandas dataframe
data = pd.read_json(r.text)
data
```

```python
#Dropping unnecessary columns
data = data.drop(columns=['success', 'timestamp', 'base', 'date'])
data
```

```python
#Converting to CSV
data.to_csv("exchange_rates.csv")
```
